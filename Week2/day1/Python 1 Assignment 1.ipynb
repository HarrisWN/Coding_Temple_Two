### Task 1: Cube Number Test
Print out all cubed numbers, up to the total value of 1000. If the cubed number is over 1000, break the loop.

Save all numbers to a list named `cube`

```python
cube = []
# YOUR CODE HERE
#Understand 
#make a list of all the cubed numbers until the solution is over 1000

#Plan
# Cube base number
#  if a solution <= 1000 
#append to list
#break loop else: break look

#Execute

for i in range(1, 1000):
    cubed = i ** 3
    if cubed <= 1000:
        cube.append(cubed)
    else:
        break


cube

#cube = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]



```

### Task 3: User Inputs
Take an input from a user for their age and convert it into an integer type. Save this to a variable to the name age. If the user is younger than 18, print `kids`, if they are 18-65, print `adults`, else, print `seniors`.

```python
#understand
# get user input as integer ("age?")
#set values for groups

#plan
#input function
#redefine age  as integer using int(age_inputs)
#if input is > 18 print = kids
#elif > 65 = adults
#else seniors


age_input = input("age?:")
age = int(age_input)

if age <18:
    print ("kids")
elif age >=18 and age <= 65:
    print("adults")
elif age > 65:
    print("seniors")
else:
    print("incorrect")
```
